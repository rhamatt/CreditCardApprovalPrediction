```python
def missing_data(final):
    miss1 = pd.DataFrame(final.isnull().sum()).reset_index()
    miss1.columns = ['Kolom', 'Jumlah']
    miss2 = pd.DataFrame(final.isnull().sum()/final.shape[0]*100).reset_index()
    miss2.columns = ['Kolom', 'Persen']
    miss = miss1.merge(miss2, how='inner', on='Kolom')
    miss = miss[miss['Jumlah']!=0]
    kolom_miss = list(miss['Kolom'])
    tipe = []
    data = []
    for kolom in kolom_miss:
        tipe.append(final[kolom].dtype)
        data.append(final[kolom][100])
    miss['Tipe'] = tipe
    miss['Contoh'] = data
    return miss
```

### 2. Memisahkan Data Train dan Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

### SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
#handle imbalance class using oversampling minority class with smote method
os = SMOTE(sampling_strategy='minority',random_state = 123,k_neighbors=5)
X_train_smote,Y_train_smote = os.fit_resample(X_train,Y_train)
X_train_smote = pd.DataFrame(data = X_train_smote,columns=X_train.columns)
Y_train_smote = pd.DataFrame(data = Y_train_smote)
```

### 4. Training Model (Dengan data Train)

```python
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix, roc_auc_score, accuracy_score
```

### SVM 

```python
from sklearn.svm import SVC

svc = SVC()
param_grid = {
    "C" : [0.1, 0,8, 1],
    "gamma" : [1, 0.1],
    "kernel" : ["linear", "rbf"]
}
```

```python
from sklearn.model_selection import GridSearchCV
CV_svc = GridSearchCV(estimator=svc, param_grid=param_grid, cv=2, refit = 'recall_score', n_jobs = -1)
CV_svc.fit(X_train_smote, Y_train_smote)
```

```python
CV_svc.best_params_
```

```python
print('Support Vector Model Accuracy : ', CV_svc.score(X_test, Y_test)*100, '%')

prediction_svc = CV_svc.predict(X_test)
print('\nConfusion matrix :')
print(confusion_matrix(Y_test, prediction_svc))
      
print('\nClassification report:')      
print(classification_report(Y_test, prediction_svc))
```

```python
plot_confusion_matrix(CV_svc, X_test, Y_test)
```
